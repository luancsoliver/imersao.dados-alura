###Desafio 07

```python
import matplotlib.pyplot as plt
dados['tempo'].value_counts().sort_index().plot.bar(color=['red','blue','green'],rot=0)
plt.xlabel('Tempo', fontsize = 12)
plt.ylabel('Quantidade de amostras', fontsize = 12)
plt.title('Quantidades de amostras por tempo de resposta', fontsize = 18)
plt.show()
```

```python
dados['tratamento'].value_counts().plot.bar(color=['red','green'],rot=0)
plt.xlabel('Tipo de tratamento', fontsize = 14)
plt.ylabel('Quantidade de amostras', fontsize = 14)
plt.title('Quantidades de amostras por tipo de tratamento', fontsize=18)
plt.show()
```
